```python
a = [-3,5,7,11,13,15,17]
b = [10,2,5,6,8,7,4]
fig = plt.figure(figsize=(5,5))
plt.plot(a,b,'r*-')
plt.xlim(-5,20)
plt.ylim(-5,20)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
for i,j in zip(a,b):
  plt.text(i,j,'({},{})'.format(i,j))
plt.title('Line Chart', size=20)
plt.suptitle('Visualization in Python', size=30, y=1.05)
```

```python
import numpy as np
print(np.pi)
print(2*np.pi)
```

```python
x=np.linspace(0,2*np.pi,400)
y=np.sin(x**2)
print(x)
print(y)
print(len(x))
print(len(y))
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('One plot using subplots')
```

```python
fig, ax = plt.subplots(4)
plt.subplots_adjust(top=1.5, right=1.5, bottom=0.1, left=0.1, hspace=0.5, wspace=0.5)
ax[0].plot(x,y)
ax[0].set_title('X,Y')
ax[1].plot(-x, y)
ax[1].set_title('-X,Y')
ax[2].plot(x, -y)
ax[2].set_title('X,-Y')
ax[3].plot(-x, -y)
ax[3].set_title('-X,-Y')
```

```python
fig, ax = plt.subplots(2,2)
plt.subplots_adjust(top=1.5, right=1.5, bottom=0.1, left=0.1, hspace=0.5, wspace=0.5)
ax[0,0].plot(x,y,'red')
ax[0,0].set_title('X,Y')
ax[0,0].set_xlabel('positive values of np.linspace(0,np.pi,400)')
ax[0,0].set_ylabel('positive values of np.sin(x**2)')
ax[0,1].plot(-x, y,'blue')
ax[0,1].set_title('-X,Y')
ax[0,1].set_xlabel('negative values of np.linspace(0,np.pi,400)')
ax[0,1].set_ylabel('positive values of np.sin(x**2)')
ax[1,0].plot(x, -y,'green')
ax[1,0].set_title('X,-Y')
ax[1,0].set_xlabel('positive values of np.linspace(0,np.pi,400)')
ax[1,0].set_ylabel('negative values of np.sin(x**2)')
ax[1,1].plot(-x, -y,'black')
ax[1,1].set_title('-X,-Y')
ax[1,1].set_xlabel('negative values of np.linspace(0,np.pi,400)')
ax[1,1].set_ylabel('negative values of np.sin(x**2)')

#for axs in ax.flat:
#  axs.set(xlabel= 'np.linspace(0,np.pi,400)', ylabel= 'np.sin(x**2)')
```

```python
fig, ax = plt.subplots(2,2)
plt.subplots_adjust(top=1.5, right=1.5, bottom=0.1, left=0.1, hspace=0.5, wspace=0.5)
ax[0,0].plot(x,y,'red')
ax[0,0].set_title('X,Y')
ax[0,1].plot(-x, y,'blue')
ax[0,1].set_title('-X,Y')
ax[1,0].plot(x, -y,'green')
ax[1,0].set_title('X,-Y')
ax[1,1].plot(-x, -y,'black')
ax[1,1].set_title('-X,-Y')

for axs in ax.flat:
  axs.set(xlabel= 'np.linspace(0,np.pi,400)', ylabel= 'np.sin(x**2)')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
sn.get_dataset_names()
```

```python
tips = sn.load_dataset('tips')
tips
```

```python
sn.stripplot(y='total_bill', data=tips, jitter=True)
plt.title('Stripplot for Total Bill')
plt.ylabel('Total Bill')
```

```python
sn.stripplot(y='total_bill', data=tips, jitter=False)
```

```python
sn.stripplot(y='total_bill', data=tips, jitter=True, size= 3, color='green')
plt.title('Stripplot for Total Bill')
plt.ylabel('Total Bill')
```

```python
sn.stripplot(x ='time' ,y='total_bill', data=tips, jitter=True, size= 3, hue='time')
plt.title('Stripplot for Total Bill Meal Type Wise')
plt.xlabel('Meal Type')
plt.ylabel('Total Bill')
```

```python
ax=sn.stripplot(x ='time' ,y='total_bill', data=tips, jitter=True, size= 3, hue='time',
             palette={'Lunch':'Yellow', 'Dinner':'Red'})
sn.move_legend(ax,'upper left')
plt.title('Stripplot for Total Bill Meal Type Wise')
plt.xlabel('Meal Type')
plt.ylabel('Total Bill')
```
