```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
Year = np.arange(2010,2025)
print(Year)
print(len(Year))
Min_Temp = np.random.randint(22,26,15)
print(Min_Temp)
print(len(Min_Temp))
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(Year,Min_Temp)
plt.xlabel('Year', size = 15)
plt.ylabel('Minimum Temperature')
plt.xlim(2008,2026)
plt.ylim(20,27)
plt.title('Line Graph Showing Minimum Temperature', size = 15)
plt.suptitle('Visualization', y=1, size = 20)
```

```python
cars=pd.read_csv('mtcars.csv')
cars
cars_float=[]
for i in cars.columns:
  if cars[i].dtype==float:
    cars_float.append(i)
cars_float
```

```python
fig = plt.figure(figsize=(12,8))
fig, axs = plt.subplots(5,4)
plt.suptitle('Scatter plots', size=20,y=1)
for i in cars_float:
  for j in cars_float:
    for m in range(5):
      for n in range(4):
        if i!=j:
          print(axs[m,n].scatter(cars[i], cars[j]))
```
