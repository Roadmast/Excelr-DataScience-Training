```python
import numpy as np
year = np.arange(2000,2025)
print(year)
temp = np.random.randint(20,30,25)
print(temp)
```

```python
fig = plt.figure(figsize= (8,5))
plt.plot(year,temp)
plt.xlim(1999,2025)
plt.ylim(15,35)
plt.xlabel('Year', size= 15)
plt.ylabel('Average Temperature', size=15)
plt.title('Comparison of Yearly Average Temperature', size = 20)
plt.suptitle('A Line Plot', size = 25, y= 1)
```

```python
fig = plt.figure(figsize= (8,5))
plt.plot(year,temp, 'r*-')
plt.xlim(1999,2025)
plt.ylim(15,35)
plt.xlabel('Year', size= 15)
plt.ylabel('Average Temperature', size=15)
plt.title('Comparison of Yearly Average Temperature', size = 20)
plt.suptitle('A Line Plot', size = 25, y= 1)
```
